ImageAI is a Python library built to empower developers to build applications and systems with self-contained deep learning and Computer Vision capabilities using a few lines of straight forward code. 

```python
from imageai.Detection import ObjectDetection
```

```python
#Let's specify the path from our input image, output image, and model.
#We have made three folders to put in in regards to this.
model_path = "C:/Users/GGPC/Desktop/object detection/model/resnet50_coco_best_v2.1.0.h5"
input_path = "C:/Users/GGPC/Desktop/object detection/input/image1.jpg"
#after output specify any random name
output_path = "C:/Users/GGPC/Desktop/object detection/output/image_new_retinanet.jpg"
```

```python
#Create an instance of the class ObjectDetection
detector1 = ObjectDetection()

#Three seperate pre-trained models to choose from 
#setModelTypeAsRetinaNet(), 
#setModelTypeAsYOLOv3(), 
#setModelTypeAsTinyYOLOv3()
detector1.setModelTypeAsRetinaNet()

#function accepts a string which contains the path to the pre-trained model:
detector1.setModelPath(model_path)

#Itloads the model from the path specified above using the setModelPath() class method.
detector1.loadModel()

#After loading the model we can specify the objects we want to be targetted.
custom = detector1.CustomObjects(person=True)

#Specify ipnut and output path so the ai can do its work.
detection1 = detector1.detectObjectsFromImage(input_image=input_path, output_image_path=output_path)
```
